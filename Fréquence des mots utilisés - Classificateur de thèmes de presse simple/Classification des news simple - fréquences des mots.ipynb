```python
import itertools


def Search_words(text,liste_mots_clés):
  liste_finale=[]
  liste=[]
  for mot_cle in liste_mots_clés:
    liste.append(re.findall(mot_cle,text))
    
  return list(itertools.chain.from_iterable(liste))

dfnews2['mots_économie']=dfnews2['textURL'].apply(lambda x: Search_words(str(x),mots_économie))
dfnews2['len_économie']=dfnews2['mots_économie'].apply(lambda x: len(x))

dfnews2['mots_écologie']=dfnews2['textURL'].apply(lambda x: Search_words(str(x),mots_écologie))
dfnews2['len_écologie']=dfnews2['mots_écologie'].apply(lambda x: len(x))

dfnews2['mots_politique']=dfnews2['textURL'].apply(lambda x: Search_words(str(x),mots_politique))
dfnews2['len_politique']=dfnews2['mots_politique'].apply(lambda x: len(x))

dfnews2['mots_sport']=dfnews2['textURL'].apply(lambda x: Search_words(str(x),mots_sport))
dfnews2['len_sport']=dfnews2['mots_sport'].apply(lambda x: len(x))

dfnews2['mots_santé']=dfnews2['textURL'].apply(lambda x: Search_words(str(x),mots_santé))
dfnews2['len_santé']=dfnews2['mots_santé'].apply(lambda x: len(x))

dfnews2['mots_technologie']=dfnews2['textURL'].apply(lambda x: Search_words(str(x),mots_technologie))
dfnews2['len_technologie']=dfnews2['mots_technologie'].apply(lambda x: len(x))

dfnews2['mots_culture']=dfnews2['textURL'].apply(lambda x: Search_words(str(x),mots_culture))
dfnews2['len_culture']=dfnews2['mots_culture'].apply(lambda x: len(x))
```

```python
dfnews2['catégorie_prédite'] = ""
for r in range(len(dfnews2['len_économie'])):
  max=np.max([dfnews2['len_politique'][r],dfnews2['len_sport'][r],dfnews2['len_écologie'][r],dfnews2['len_économie'][r],dfnews2['len_santé'][r],dfnews2['len_technologie'][r],dfnews2['len_culture'][r]])
  if(dfnews2['len_économie'][r]==max):
    res='Economie'
  if(dfnews2['len_écologie'][r]==max):
    res='Ecologie'
  if(dfnews2['len_politique'][r]==max):
    res='Politique'
  if(dfnews2['len_sport'][r]==max):
    res='Sport'
  if(dfnews2['len_santé'][r]==max):
    res='Santé'
  if(dfnews2['len_technologie'][r]==max):
    res='Technologie'
  if(dfnews2['len_culture'][r]==max):
    res='Culture'
  if(max==0):
    res='Faits divers'
  dfnews2['catégorie_prédite'][r]=res
```
